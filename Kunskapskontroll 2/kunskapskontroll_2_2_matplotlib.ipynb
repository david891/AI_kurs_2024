### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

* In the explicit (object oriented) syntax you use the axes object(s) when you plot.
* In the implicit (procedural) syntax matplotlib assumes which axes to use when you plot.
* The recommended way is to use the explicit syntax.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

* Figure: The container of one or more plots (axes).
* Axes: The canvas on which plotting is done. Several plots can share the same axes and overlap on the canvas.
* Axis: The x-axis and y-axis with ticks, tick labels and axis labels to show information on what the plot shows.
* Artist: Everything that you se on the canvas is an artist. The figure, axes and axis are artists. And the lables, lines, dots, bars, text etc. that are drawn on the axes. All of that are artists that take care of the rendering in the background.

### When plotting in Matplotlib, what is the expected input data type?

* The expected input data typ when plotting in Matplotlib is a list or array of some sort. A python list, a numpy array or a pandas series works.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x  = np.linspace(-4,4,81)
y = x**2
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
x  = np.linspace(-4,4,81)
y = x**2
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fix, axs = plt.subplots(1,2, figsize=(10,5))
axs[0].scatter(random_data_x,random_data_y)
axs[1].bar(x=names,height=values)
```
